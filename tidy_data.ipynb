## 3. Cake Recipes

- Read the cake_recipes.csv data. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.
- Tidy the data as necessary.
- Which recipe, on average, is the best? recipe b
- Which oven temperature, on average, produces the best results? 275
- Which combination of recipe, rack position, and temperature gives the best result? recipe b, bottom rack, 300 degrees

```python
cake_df = pd.read_csv('cake_recipes.csv')
cake_df
```

```python
# going to try the split() function. What to return it into those? Can I use the creating new columns, or should I create a new dataframe and simple concat that dataframe? I think I'll try the latter.
```

```python
r_p = cake_df['recipe:position'].str.split(':', expand=True)
r_p
```

```python
# Testing to see if the column name is a str or an int (expecting an int), so that I can input the correct type into the rename function.
# I could also skip this step and simply find out experimentally by running the rename() function until it works, but I'm curious to try it this way.

type(r_p.columns[0])
```

```python
# The column titles are ints (as expected), so I'll use an int 0, 1 in my rename function:

r_p.rename(columns = {0: 'recipe', 1: 'position'}, inplace = True)
r_p
```

```python
# Concat the two dataframes, and drop the old recipe:position column:

cake_df = pd.concat([r_p, cake_df], axis=1).drop(columns = 'recipe:position')
cake_df
```

```python
# Now a have a slightly better df. But too wide still.
cake_df
```

```python
# Use the melt function:

cake_df = cake_df.melt(id_vars=['recipe', 'position'])
cake_df
```

```python
cake_df.rename(columns = {'variable': 'temperature', 'value': 'total_score'}, inplace=True)
cake_df
```

```python
# Grouping by the recipe since we want to know which recipe has the highest average (mean) score.

cake_df.groupby('recipe').mean()

# Recipe b has the highest overall average score
```

```python
# I can do the same thing, except grouping by the temprature:

cake_df.groupby('temperature').mean()

# Temp set at 275 has the highest overall average score
```

#### Which combination of recipe, rack position, and temperature gives the best result?

```python
cake_df.pivot_table('total_score', ['recipe', 'position', 'temperature'])
```

```python
# Idenfiying the location (the index) of the highest score, and then narrowing down on just that row (observation)
# By using the brackets on the dataframe, I am iterating through the dataframe to find the matching index (I believe that's how it's working)

index_of_highest_score = cake_df.total_score.argmax()
cake_df[cake_df.index == index_of_highest_score]
```

```python
# Or, using iloc:

cake_df.iloc[cake_df.total_score.argmax()]
```
